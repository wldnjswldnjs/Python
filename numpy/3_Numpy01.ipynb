## logspace()
- 로그 스케일의 linspace
- 로그스케일로 지정된 범위에서 num 개수 만큼 균등 간격으로 데이터 생성한 후 배열 만듬
- logspace( a, b, c)일 경우,
- log^a와 log^b 사이에 등간격인 c개의 행벡터를 생성하는 함수.

```python
a = np.logspace(1, 10, 20, endpoint=True)
```

```python
#endpoint=True 혹은 endpoint=False를 옵션으로 줄 수 있음. 
#디폴트(Default)로 True이며, 이는 stop으로 주어진 값을 포함시킬 것인지 아닌지를 선택하는 옵션임.

```

```python
a
```

## random.rand()
- 지정된 shape 배열 생성 후 난수로 초기화
- 0에서 1사이 값의 균등 분포를(uniform distribution) 가지는 값들을 추출

```python
a = np.random.rand(2,2,2)
print(a)
```

```python
plt.hist(a[1,:], bins=10)  
plt.show()
```

```python
data = np.random.rand(10000) #데이터의 개수 10000개
import matplotlib.pyplot as plt
plt.hist(data, bins=100)
plt.show()
```

## random.randn()
- shape 배열 생성 후 난수로 초기화

## 약속된 난수
- 무작위 수를 만드는 난수는 특정 시작 숫자로부터 난수처럼 보이는 수열을 만드는 알고리즘의 결과물이다. 따라서 시작점을 설정함으로써 난수 발생을 재연할 수 있다. 난수의 시작점을 설정하는 함수가 np.random.seed 이다.

```python
np.random.random((2, 2))
```

```python
np.random.randint(0, 10, (2, 3))
```

```python
np.random.seed(5)
np.random.random((2, 2))
```

```python
np.random.random((2, 2))
```

```python
np.random.seed(5)
np.random.randint(0, 10, (2, 3))
```
