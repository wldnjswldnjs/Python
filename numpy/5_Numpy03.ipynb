- ravel은 원본의 뷰를 반환하기 때문에 ravel로 반환된 데이터를 수정하면 원본 데이터도 수정 된다.

```python
a2 = np.random.choice(np.arange(2,22), 12).reshape(3,4) #2 이상 22 미만의 정수 중 12개 출력
a2
```

```python
np.argmin(a2, axis=1)  #가로축 요소들끼리 비교한 최소값
```

# 3.배열복사
- ndarray 배열 객체에 대한 slice, indexing이 반환하는 배열은 새로운 객체가 아닌 기존 배열의 뷰(view)
- 반환된 배열의 값을 변경하면 원본 배열에 반영된다.
- 기본 배열로부터 새로운 배열을 생성하기 위해서는 copy함수를 통해 명시적으로 사용해야한다.
- 원본 배열과 copy 함수로 복사된 사본 배열은 완전히 다른 별도의 객체이다.

## (1) copy() 사용 안했을 때

```python
a1 = np.arange(1,21).reshape(4,5)
a1
```

```python
a2 = a1[:, :2]   
a2               #원본, 즉 a1의 view
```

```python
a2[0,1] = 3
a2
```

```python
a1      #변경 값이 원본에 반영됨.
```

## (2) copy() 사용했을 때 

```python
a1 = np.arange(1,21).reshape(4,5)  
a1
```

```python
a2 = np.copy(a1[:, :2])     #새로운 객체 생성
a2
```

```python
a2[0,1] = 3
a2
```

```python
a1    #원본의 값이 변하지 않음.
```

```python
a2
```

## (1) resize
- reshape와 같이 배열의 shape를 변경하는 부분에서는 유사하다.
- reshape는 배열의 요소 수를 변경할 수 없다.
- resize는 shape 변경 가능하며, 배열 요소 수도 변경 가능하다.

```python
import numpy as np
```

- reshape

a1.reshape 와 np.reshape 


```python
a1 = np.arange(1,13).reshape(2,6)
a1                         # 배열의 요소 수는 12개
```

```python
np.reshape(a1, (3,4))
```

```python
a1
```

```python
a1.reshape(3,4)     
```

```python
a1       #원본의 배열이 바뀌지 않음
```

```python
a1.reshape(3,3)     #2차원 3행 3열, 즉 요소 수가 9개인 배열로 변경 -> 불가함. 
```

- resize

a1.resize 와 np.resize

```python
a1 = np.arange(1,13).reshape(2,6)
a1                         # 배열의 요소 수는 12개
```

```python
np.resize(a1, (6,2))  
```

```python
a1
```

```python
a1.resize((6,2))    
```

```python
a1    #앞의 변수를 객체로 해서 사용했을 때 원본의 배열이 바뀜
```

```python
np.resize(a1, (3,3))  #요소의 개수가 12개인 배열에서 9개인 배열로 변경
```

- reshape의 copy와 resize의 copy

```python
a1 = np.arange(1,13).reshape(2,6)
a1
```

```python
b1 = np.reshape(a1, (2,6))
b1[1,0] =100
b1
```

```python
a1                            #a1과 b1은 이름만 다를 뿐 같은 메모리 공간을 공유함
```

```python
a1 = np.arange(1,13).reshape(2,6)
a1
```

```python
b1= np.reshape(a1, (2,6)).copy()
b1[1,0] =100
b1
```

```python
a1
```

```python
#resize
b1 = np.resize(a1, (2,6))
b1[1,0] =100
b1
```

```python
a1                     #resize는 reshqpe와 달리 b1을 새로운 공간을 만들어 저장함. 
```
