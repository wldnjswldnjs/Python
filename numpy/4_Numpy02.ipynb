### 3)np.where
- if 절과 같은 효과
- np.where(조건, True일 때 원하는 value, False일 때 원하는 value)

````python
a = np.array([2,3,1])
a
````

````python
np.where(a > 1, 10, 20) # True, True, False
````

````python
#조건 뒤에 오는 value가 없을 때는 True일 때의 index값 반환
np.where(a>1)  # a > 1 인 인덱스
````

## (4). NumPy 성능비교
- jupyter notebook 환경에서 코드의 성능을 볼 수 있는 %timeit 사용
- 구구단 100단 연산을 비교

### 1) list와 for사용

````python
for i in range(1,101):
    for j in range(1,101):
        print(i*j)
````

````python
list_a = range(1,101)
list_b = range(1,101)
````

````python
[i * j for i in list_a for j in list_b]
````

````python
%timeit [i * j for i in list_a for j in list_b]     #1000번 반복했을때, best 5의 평균수행시간이 647 마이크로세컨드
````

````python
%timeit()
````

### 2)numpy 사용

````python
array_a = np.arange(1,101).reshape(-1,1)
array_a
````

````python
array_b = np.arange(1,101)
array_b
````

````python
array_a * array_b  # 100by1 배열과 1by100배열의 곱셈연산 
````

````python
%timeit array_a * array_b #100000회 반복했을 때, best 5의 평균수행시간이 13.8 마이크로세컨드
````

````python
%timeit -n 1000 array_a * array_b  # 1000회 돌리기
````
