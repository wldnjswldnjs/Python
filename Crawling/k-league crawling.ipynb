```python
from bs4 import BeautifulSoup
from selenium import webdriver
from datetime import datetime, timedelta
from html_table_parser import parser_functions

import time
import pandas as pd
import os
import re
```

```python
# 3. 수집된 데이터를 저장할 파일 이름 입력
fc_name = input('3. 결과를 저장할 csv 형식의 파일명을 쓰세요 : ')
```

```python
## 크롬 드라이버 설정 및 웹 페이지 열기
chrome_path = '/temp/chromedriver103/chromedriver.exe'
driver = webdriver.Chrome(chrome_path)

date2 = []     # 경기 날짜
time2 = []     # 경기 시간 
round2 = []    # 경기 라운드
stadium2 = []  # 경기장
team_a2 = []    # 팀A
team_b2 = []    # 팀B
a_result2 = []  # 팀A_결과
b_result2 = []  # 팀B_결과

url = 'https://sports.news.naver.com/kfootball/schedule/index?'

year1 = '2019'
month1 = '3'
a = int(month1)

for i in range(a-1,11):
    i += 1
    
    full_url = url + 'year=' + year1 + '&month=' + str(i) + '&category=kleague2'

    driver.get(full_url)
    time.sleep(5)

    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
# print(start_date2[s_mon])
# print(end_date2[e_mon])
    

    soup1 = soup.find_all('tr')

    time.sleep(5)

    for b in soup1:
        try:
            hour = b.find('span', 'time').get_text()
    #                 a = b.find('span', 'year').get_text()
    #                 b = b.find('div', 'inner').find('em').get_text()
        except:
            continue
        else:
    #                 string = a + b
    #                 numbers = re.sub(r'[^0-9]', '-', string )
    #                 numbers = datetime.strptime(numbers,'%Y-%m-%d')
    #                 print(numbers)

    #                 hour = b.find('span', 'time').get_text()
            print('경기 시간:', str(hour))
            time2.append(hour)

            round1 = b.find('td', 'game_content last').find('span').get_text()
            print('경기 라운드: %s' % round1 )
            round2.append(round1)

            stadium = b.find('span', 'place').get_text()
            print('경기장 : %s' % stadium)
            stadium2.append(stadium)

            team_a = b.find('span', 'team_left').find('span', 'name').get_text()
            print('팀 A :%s' % team_a)
            team_a2.append(team_a)

            team_b = b.find('span', 'team_right').find('span', 'name').get_text()
            print('팀 B :%s' % team_b)
            team_b2.append(team_b)

            a_result = b.find('span', 'team_left').find('span', 'score').get_text()
            print('팀A_결과 :%s' % a_result)
            a_result2.append(a_result)

            b_result = b.find('span', 'team_right').find('span', 'score').get_text()
            print('팀B_결과 :%s' % b_result)
            b_result2.append(b_result)


            print('\n')
```
