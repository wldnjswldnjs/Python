### 5. 변동률 계산

```python
# 2019년 대비 2022년 변동률
```

```python
# 비교 시점 : 2022년
# 기준 시점 : 2019년
```

```python
data = data.transpose()
```

```python
data
```

```python
# 2019년도 데이터
data['2019년 01월']
```

```python
# 2022년도 데이터
data['2022년 01월']
```

```python
data_v = (data['2022년 01월'] - data['2019년 01월']) / data['2019년 01월']
```

```python
data_v = data_v * 100
```

```python
data_v
```

```python
data_v.plot(kind='barh')

plt.show()
```

```python
data_v = data_v.sort_values() # 오름차순
```

```python
data_v.plot(kind='barh')

plt.show() # 그래프는 아래부터 그려짐
```

```python
# 상위 3위
data_v.iloc[-3:]
```

```python
# 하위 3위
data_v.iloc[0:3]
```

```python
plt.figure(figsize = (15,5))

plt.subplot(1,2,1)
data_v.iloc[-3:].plot(kind='barh')
plt.title('2019년-2022년 매매가격지수 변동률 상위 3위')

plt.subplot(1,2,2)
data_v.iloc[:3].plot(kind='barh')
plt.title('2019년-2022년 매매가격지수 변동률 하위 3위')

plt.show()
```
